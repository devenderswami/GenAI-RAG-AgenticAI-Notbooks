```python
from llama_index.core import SimpleDirectoryReader, Settings,VectorStoreIndex
from llama_index.core.node_parser import SemanticSplitterNodeParser
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
```

```python
Settings.embed_model = HuggingFaceEmbedding("sentence-transformers/all-MiniLM-L6-v2")
Settings.llm = None 
```

```python
docs = SimpleDirectoryReader("./GenAI/GenAI-NoteBooks/coffee_pages",required_exts=[".html"]).load_data()
# ---- Semantic splitting (meaning-aware chunking) ----
sem_parser = SemanticSplitterNodeParser(
    embed_model=Settings.embed_model,
    buffer_size=3,
    breakpoint_percentile_threshold=95
)
# This returns Nodes
nodes = sem_parser.get_nodes_from_documents(docs)

# ---- Build Index using Nodes ----
index = VectorStoreIndex(nodes)

# ---- Retrieve ----
hits = index.as_retriever(similarity_top_k=3).retrieve("brewing temperature")

for h in hits:
    print("-" * 70)
    print(h.node.get_content()[:700], "...")
```
