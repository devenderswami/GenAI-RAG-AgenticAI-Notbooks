```python
import os
from pinecone import Pinecone      # if you plan to use Pinecone
from llama_index.vector_stores.pinecone import PineconeVectorStore
from llama_index.core import VectorStoreIndex
from llama_index.llms.google_genai import GoogleGenAI
from llama_index.embeddings.google_genai import GoogleGenAIEmbedding
```

```python
# --- Connect to Pinecone ---
pc = Pinecone(api_key="")
index = pc.Index("coffeeindex") # replace with your index name
```

```python
# --- Wrap Pinecone VectorStore ---
li_store = PineconeVectorStore(pinecone_index=index)
```

```python
# --- Gemini Embeddings (must match what you used for upserts) ---
embed  = GoogleGenAIEmbedding(
 model_name="models/text-embedding-004",
 api_key=""
)
```

```python
# --- Build LlamaIndex wrapper ---
index_li = VectorStoreIndex.from_vector_store(li_store,embed_model=embed)
```

```python
# --- Gemini LLM ---
li_llm = GoogleGenAI(
 model="gemini-2.5-flash",
 api_key=""
)
```

```python
# --- Query Engine ---
qe = index_li.as_query_engine(llm=li_llm)
print(qe.query("A turmeric latte, also known as "))
```
