```python
from llama_index.core import SimpleDirectoryReader, Settings,VectorStoreIndex
from llama_index.core.node_parser import SentenceSplitter
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
```

```python
Settings.embed_model =HuggingFaceEmbedding("sentence-transformers/all-MiniLM-L6-v2")
Settings.llm = None # no generation
```

```python
docs = SimpleDirectoryReader("./GenAI/GenAI-NoteBooks/coffee_pages",required_exts=[".html"]).load_data()
index = VectorStoreIndex.from_documents(docs, transformations=[SentenceSplitter(chunk_size=512, chunk_overlap=64)])
```

```python
hits = index.as_retriever(similarity_top_k=3).retrieve("brewing temperature")
for h in hits:
 print("-" * 70, "\n", h.node.get_content()[:700], "...")
```
