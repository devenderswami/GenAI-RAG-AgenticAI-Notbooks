```python
import google.generativeai as genai
```

```python
GEMINI_API_KEY= ""
```

```python
genai.configure(api_key=GEMINI_API_KEY)
```

```python
SYSTEM_PROMPT = """
you are customer support agent for ecommerce site which help the customer related queries your tone should be helpful 
"""
```

```python
model = genai.GenerativeModel(
        model_name="gemini-2.5-flash",
        system_instruction=SYSTEM_PROMPT
)
```

```python
chat = model.start_chat()
```

```python
def main():
    while True:
        msg = input("Enter you query: ").strip()
        if msg.lower() == "exit":
            print("session ended")
            break
        if not msg:
            continue
        print("chat:",chat)
        response = chat.send_message(msg).text
        print("Assistant:",response)
     
if __name__ == "__main__":
    main()
    
```
