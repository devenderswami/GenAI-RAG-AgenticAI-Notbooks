```python
prompts = [

    "Summarize the history of the internet in 3 lines.",

    "Explain what blockchain is for a beginner.",

    "What are the main components of a neural network?"]
```

```python
# Without batch!

# Invoke one-by-one

for i, prompt in enumerate(prompts):

    response = llm.invoke(prompt)

    print(f"\n Prompt {i+1}: {prompt}")

    print(f" Response: {response.content}")
```

```python
# with batch!

responses = llm.batch(prompts)

for i, response in enumerate(responses):

    print(f"\n Prompt {i+1}: {prompts[i]}")

    print(f" Response: {response.content}")

```

```python
from pydantic import BaseModel, EmailStr

from langchain.output_parsers import PydanticOutputParser

# 1. pydantic model : 

class Person(BaseModel):

    name : str

    email : str

 

# Step 2 : Create the output parser

parser = PydanticOutputParser(pydantic_object = Person)

input_text = "Hi my name is Devender,and I was born 01 jan 2000, todays date 10 jan 2021 and my email id is dev@gmail.com"

format_instruction = parser.get_format_instructions()

final_prompt = f"""Extract person info from the following text:

{input_text}

{format_instruction}

"""

final_prompt

 

# Step 4 . 

from langchain_google_genai import ChatGoogleGenerativeAI

# Initialize the model

llm = ChatGoogleGenerativeAI(

    model="gemini-1.5-flash",

    google_api_key=""

)

 

response = llm.invoke(final_prompt)

print(response.content)
```
