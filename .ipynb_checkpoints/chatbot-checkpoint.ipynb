```python
import google.generativeai as genai
```

```python
GEMINI_API_KEY= ""
```

```python
genai.configure(api_key=GEMINI_API_KEY)
```

```python
#SUMMARY_THRESHOLD â†’ after 6 messages, the bot will summarize the conversation so it doesnâ€™t forget older stuff.
MODEL = "gemini-2.5-flash"
SUMMARY_THRESHOLD = 6
```

```python
def summarize_history(history: list[dict]) -> str:
    """
    (expression for variable in iterable)
    â€¢	expression = what to do with each item (f"{m['role']}: {m['text']}")
	â€¢	variable = name for each item in the loop (m)
	â€¢	iterable = the collection youâ€™re looping over (history)

    pieces = []
    for m in history:
        pieces.append(f"{m['role']}: {m['text']}")
        transcript = "\n".join(pieces)
    
    """
    transcript = "\n".join(f"{m['role']}: {m['text']}" for m in history)
    
    prompt = f"""
      You are a conversation summarizer. Summarize the chat below, capturing main points and intent.
      {transcript}
       --
      Provide a concise summary:"""
    model = genai.GenerativeModel(MODEL)
    resp = model.generate_content(prompt)
    return resp.text.strip()
    
    
```

```python
# â€”â€”â€” Main chat loop â€”â€”â€”
def run_chat():
    history = []
    print("ðŸ§  Gemini Chat â€” type 'quit' to exit.")

    while True:
        user = input("You: ")
        if user.lower() in ("quit", "exit"):
            print("Session ended.")
            break

        history.append({"role": "user", "text": user})

        # Summarize if threshold exceeded
        if len(history) > SUMMARY_THRESHOLD:
            summary = summarize_history(history)
            history = [
                {"role": "system", "text": f"Summary of previous convo: {summary}"}
            ]
            print("\n--- [SYSTEM] Conversation summarized. â€”â€”â€”\n")

        # Build transcript from history
        transcript = "\n".join(f"{m['role']}: {m['text']}" for m in history)

        # Call Gemini model
        model = genai.GenerativeModel(MODEL)
        resp = model.generate_content(transcript + "\nassistant:")

        assistant_msg = resp.text.strip()
        print("Gemini:", assistant_msg)

        # Store assistant response
        history.append({"role": "assistant", "text": assistant_msg})


if __name__ == "__main__":
    run_chat()
```
