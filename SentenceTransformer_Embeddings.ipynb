```python
from sentence_transformers import SentenceTransformer,util
```

```python
# Example documents
docs = [
"Cats are playful and independent pets.",
"Electric cars are eco-friendly vehicles.",
"Dogs are loyal and friendly animals.",
"Kitten care includes feeding, grooming, and vaccinations."
]
```

```python
# intiate the model
model = SentenceTransformer("all-MiniLm-L6-v2")
```

```python
#Encode all documents
doc_embeddings = model.encode(docs,convert_to_tensor=True)
```

```python
#Query
query = "How do i take care of kitten"
```

```python
#encode query
query_embedding = model.encode(query,convert_to_tensor=True)
```

```python
#find most similar docs
cos_scores = util.cos_sim(query_embedding,doc_embeddings)[0]
top_results = cos_scores.argsort(descending=True)[:2]
```

```python
for idx in top_results:
  print(f"Score:{cos_scores[idx]:.4f} Document: {docs[idx]}")
```
