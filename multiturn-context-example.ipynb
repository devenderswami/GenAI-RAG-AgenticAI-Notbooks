```python
from openai import OpenAI
```

```python
client = OpenAI(api_key = '',base_url='https://generativelanguage.googleapis.com/v1beta/openai/')
```

```python
messages = [
     {"role":"system", "content" : "You are a career counsellor. Give clear, empathetic, helpful reponses and ask follow-up questions."},
    {"role":"user", "content" : "Hi, I'm Nikhil. I enjoy math and programming and I just completed my B.Sc."}
]
```

```python
response = client.chat.completions.create(
    model = 'gemini-2.5-flash',
    messages = messages
)
```

```python
reply = response.choices[0].message.content
```

```python
print(reply)
```
