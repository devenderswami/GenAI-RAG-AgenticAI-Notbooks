```python
import os
os.environ["SERPAPI_API_KEY"] = ""
from crewai import LLM, Agent, Task, Crew, Process
from crewai_tools import SerpApiGoogleSearchTool
os.environ["GOOGLE_API_KEY"] = ""
my_llm = LLM(model="gemini/gemini-2.5-flash", temperature=0.2)
print("âœ… CrewAI Installed.")
```

```python
# --- 2. DEFINE AGENTS ---
search_tool = SerpApiGoogleSearchTool()
# Specialist: Researcher (Equipped with the Tool)
researcher = Agent(
    role='Lead Researcher',
    goal='Find real-time robotics trends.',
    backstory="You are a research specialist who searches the web for verifyable facts.",
    tools=[search_tool], # <--- REAL TOOL ASSIGNED HERE
    llm=my_llm,
    verbose=True
)

# Specialist: Writer (No tools, just writing skills)
writer = Agent(
    role='Content Strategist',
    goal='Write content.',
    backstory="You transform technical research into engaging fun summaries.",
    llm=my_llm,
    verbose=True
)
```

```python
# Explicit Manager
project_manager = Agent(
    role = 'Project Manager',
    goal = 'Coordinate the robotics report',
    backstory = 'You oversee the project and delegate tasks to your team',
    llm=my_llm,
    verbose=True,
    allow_delegation = True
)
```

```python
# Task : assigned to the project manager
main_task = Task(
    description = "Search for the latest robotics trends from the last 6 months and write a summary.",
    expected_output = "A professional executive summary with real sources.",
    agent = project_manager
)
```

```python
federated_crew = Crew(
    agents = [researcher,writer],
    tasks  = [main_task],
    process = Process.hierarchical,
    manager_agent = project_manager, # explicit manager
    verbose=True
)
```

```python
result = federated_crew.kickoff()
```
