```python
import os
os.environ["SERPAPI_API_KEY"] = ""
from crewai import LLM, Agent, Task, Crew, Process
from crewai_tools import SerpApiGoogleSearchTool
os.environ["GOOGLE_API_KEY"] = ""
my_llm = LLM(model="gemini/gemini-2.5-flash", temperature=0.2)
print("âœ… CrewAI Installed.")
```

```python
search_tool = SerpApiGoogleSearchTool()

# 1. Agents
researcher = Agent(
    role="Senior Market Researcher",
    goal="Fetch latest market news and current stock price",
    tools=[search_tool],
    llm=my_llm,
    verbose=True,
    backstory = "You are an expert in Stock Market Research"
)

risk_officer = Agent(
    role="Risk Officer",
    goal="Analyze risk and give Buy/Hold/Sell verdict",
    llm=my_llm,
    verbose=True,
    backstory = "You are Risk Officer who help in analysing risks and give verdict of buy/hold/sell"

)
# 2. Tasks
research_task = Task(
    description="Find latest news and stock price for {company}. Only factual data.",
    expected_output="Bullet points with headlines and current price.",
    agent=researcher
)

risk_task = Task(
    description="Analyze the research and give a Buy/Hold/Sell recommendation.",
    expected_output="One cohesive paragraph with final verdict.",
    agent=risk_officer,
    context = [research_task]
)
```

```python
crew = Crew(
    agents=[researcher, risk_officer],
    tasks=[research_task, risk_task],
    process=Process.sequential,
    verbose=True
)
```

```python
print("Kickoff")
result = crew.kickoff(inputs={'company':'Google'})
print(result)
```
