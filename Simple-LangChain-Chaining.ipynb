```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash",google_api_key="",temperature=0)
```

```python
# --- STEP 1: The Creative Drafter ---
draft_prompt = PromptTemplate.from_template(
        "Write a short, enthusiastic blog post about: {topic}. "
        "Use lots of emojis and slang!"
    )
draft_chain = draft_prompt | llm | StrOutputParser()
 # --- STEP 2: The Strict Editor ---
editor_prompt = PromptTemplate.from_template(
        """
        You are a strict technical editor. Rewrite the following text.

        Constraints:
        1. Remove all emojis.
        2. Change the tone to be professional and academic.
        3. Do not mention that you are an AI.

        Text to rewrite:
        {draft}
        """
    )
editor_chain = editor_prompt | llm | StrOutputParser()
# --- EXECUTION ---
topic = "The importance of Python in AI"
print("Generating Draft...")
try:
 draft = draft_chain.invoke({"topic": topic})
 print(f"\n--- ğŸ“ STEP 1 OUTPUT (DRAFT) ---\n{draft}\n")

 print("Refining Draft...")
 final_version = editor_chain.invoke({"draft": draft})
 print(f"\n--- âœ… STEP 2 OUTPUT (FINAL) ---\n{final_version}")
except Exception as e:
 print(f"An error occurred: {e}")
```
