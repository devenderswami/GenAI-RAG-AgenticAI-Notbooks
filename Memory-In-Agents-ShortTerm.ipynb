```python
# ============================================================
# STEP 2: Import Dependencies and Setup API Key
# ============================================================
import os
from getpass import getpass
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import ConversationChain
from langchain.memory import (
    ConversationBufferMemory,
    ConversationSummaryMemory,
    ConversationBufferWindowMemory,
    ConversationSummaryBufferMemory,
)

os.environ["GOOGLE_API_KEY"] = ""

# Initialize Gemini model (temperature=0 for consistent responses)
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
print(llm)
```

```python
# EXAMPLE 1: Conversation Buffer Memory
# Stores - all messages - good for short conversations

buffer_memory = ConversationBufferMemory()

conv_chain = ConversationChain(
    llm=llm,
    memory=buffer_memory,
    verbose = True # shows the prompts sent to LLMs (step by step on whats happening)
)

print(conv_chain.invoke({"input": "Hi there, I'm Akash"})["response"])
```

```python
print(conv_chain.invoke({"input": "What should be linkedin profile name? And Can you help me with some content for my first post on LinkedIn?"})["response"])
```

```python
print(conv_chain.invoke({"input": "Ok, similarily what should my funny twitter user name based on my actual name?"})["response"])

# Notice: Each turn includes ALL previous messages in the prompt
```

```python
# ConversationSummaryMemory
# Summarizes conversation instead of storing everything


summary_memory = ConversationSummaryMemory(llm=llm)

conv_chain_summary = ConversationChain(
    llm=llm,
    memory=summary_memory,
    verbose = True # shows the prompts sent to LLMs (step by step on whats happening)
)

print(conv_chain_summary.invoke({"input": "Hi there, I'm Akash"})["response"])
```

```python
print(conv_chain_summary.invoke({"input": "Can you help me with writing a funny tweet?"})["response"])
```

```python
print(conv_chain_summary.invoke({"input": "india k bare mey batye in 2 lines?"})["response"])
```

```python
# Example Conversation Buffer windo Memory
# Only Keeps last k interactions - good for medium conversations

window_memory = ConversationBufferWindowMemory(k=2)

conv_chain_window = ConversationChain(
    llm=llm,
    memory=window_memory,
    verbose = True # shows the prompts sent to LLMs (step by step on whats happening)
)
print(conv_chain_window.invoke({"input": "Hi there, I'm Devender"})["response"])
```

```python
print(conv_chain_window.invoke({"input": "can ou write me funny linked in memory"})["response"])
```

```python
print(conv_chain_window.invoke({"input": "Explain llm to a 5y old in 5 lines?"})["response"])
```

```python
print(conv_chain_window.invoke({"input": "Tell me a fuuny joke about pm modi?"})["response"])
```

```python
print(conv_chain_window.invoke({"input": "Tell me a fuuny joke about rahul gandhi?"})["response"])
```

```python
# EXAMPLE 4: Conversation Summary Buffer Memory
# Hybrid : Keeps the recent messages + summary of older ones

summary_buffer_memory = ConversationSummaryBufferMemory(
    llm = llm,
    max_tokens = 200) # when this limit is exceeds, older message will be summarized


conv_chain_buff_summary = ConversationChain(
    llm=llm,
    memory=summary_buffer_memory, verbose = True)

print(conv_chain_buff_summary.invoke({"input": "Im Akash, Can you write me something about AI agents, keep it short"})["response"])
```

```python
print(conv_chain_buff_summary.invoke({"input": "What is the diff between Agents v/s agentic AI in 2 lines? "})["response"])
```

```python
print(conv_chain_buff_summary.memory.moving_summary_buffer)
```

```python
# ============================================================
# SUMMARY: When to Use Each Memory Type
# ============================================================
print("\n" + "="*60)
print("WHEN TO USE EACH MEMORY TYPE:")
print("="*60)
print("""
1. ConversationBufferMemory
   - Use: Short conversations (< 10 exchanges)
   - Pros: Full context always available
   - Cons: Can exceed token limits quickly

2. ConversationSummaryMemory
   - Use: Long conversations where you need the gist
   - Pros: Handles unlimited conversation length
   - Cons: Loses specific details, requires extra LLM calls

3. ConversationBufferWindowMemory
   - Use: Medium conversations, recent context important
   - Pros: Balanced approach, controlled token usage
   - Cons: Loses older context

4. ConversationSummaryBufferMemory
   - Use: Long conversations needing both detail and summary
   - Pros: Best of both worlds
   - Cons: More complex, requires tuning token limit
""")
```
